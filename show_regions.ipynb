```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(3, 3, figsize=(15,10))
regions = [f'x{i :02d}' for i in range(1, 11) if i != 5]
row = 0
col = 0
for region in regions:
    truth = np.load(f'data/truth_masks/truth_{region}.npy')
    ax[row, col].imshow(truth)
    ax[row, col].set_title(region)
    col += 1
    if col < 3:
        continue
    else:
        col = 0
        row += 1
```
